# Import Data

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/Users/kellyliu/Documents/00. McGill MMA/## INSY Text Analytics/Group Final/user_segmentation.csv')

# Display the first 10 rows
df.head(30)
```

## lowercase all words except header

```python
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

# Sentiment Analysis on POS with Vader

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

# Load the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Perform sentiment analysis on segment 0
df['sentiment_scores_pos_vader'] = df['attributes_patterns'].apply(lambda x: sia.polarity_scores(x))
df['sentiment_score_pos_vader'] = df['sentiment_scores_pos_vader'].apply(lambda x: x['compound'])

df.head(10)
```

# Sentiment Analysis on Review with SpaCy

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load the SpaCy model
nlp = spacy.load('en_core_web_sm')

# Add the SpaCyTextBlob component to the pipeline
nlp.add_pipe('spacytextblob')

# Perform sentiment analysis on the "review" column
df['sentiment_score_review_spacy'] = df['review'].apply(lambda x: nlp(str(x))._.polarity)

# Perform sentiment analysis on the "attributes_patterns" column
df['sentiment_score_pos_spacy'] = df['attributes_patterns'].apply(lambda x: nlp(str(x))._.polarity)

df.head(10)


```

```python
df.to_csv('POS Vader vs. Review Spacy.csv', index=False)
```

```python
# Split the CSV into two segments based on the cluster column
segment_0 = df[df['cluster'] == 0]
segment_1 = df[df['cluster'] == 1]

# Output the segments as separate CSV files
segment_0.to_csv('segment_0.csv', index=False)
segment_1.to_csv('segment_1.csv', index=False)
```
